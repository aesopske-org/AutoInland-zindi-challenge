```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
```

```python
# read the data
os.listdir("C:\\Users\\Maria\\Videos\\thende\\data")
df = pd.read_csv('clean_data/cleaned_train_included_color_droped_nulls.csv')
```

```python
# we need to convert the non numeric columns to numeric
from sklearn.preprocessing import OrdinalEncoder
ord_enc = OrdinalEncoder()
import numpy as np

# convert gender to numeric
non_numerals = ["Gender","Car_Category","Subject_Car_Make","LGA_Name","State","ProductName"]
for i in non_numerals:
    ord_enc = OrdinalEncoder()
    df[i] = ord_enc.fit_transform(np.array(df[i]).reshape(-1, 1))

# create features from the date columns
df["First Transaction Date"] = pd.DatetimeIndex(df["First Transaction Date"])
df["transaction_month"] = [i.month for i in df["First Transaction Date"]]

df.head()
```

# Corelations 

* appeared to have very low corelations between the target and all the variable and where weak corelation is present its negative

```python
# the first multivariate analysis is the corelation test
sns.color_palette("icefire", as_cmap=True)
sns.heatmap(df.corr(), cmap="PiYG")
```
